## 2. Converting Addresses to Coordinates

Section 15, Video 127.

Tool: [GeoPy](https://geopy.readthedocs.io/en/stable/#)

```python
df9 = pd.read_csv("supermarkets.csv")
df9
```

```python
from geopy.geocoders import ArcGIS
```

```python
nom = ArcGIS()
```

```python
n = nom.geocode("3995 23rd St, San Francisco, CA 94114")
```

```python
n.longitude
```

```python
n.latitude
```

```python
type(n)
```

```python
df9["Full_Addr"] = df9["Address"] + ", " + df9["City"] + ", " + df9["State"] + ", " + df9["Country"]
df9
```

```python
df9["Coordinates"] = df9["Full_Addr"].apply(nom.geocode)
df9
```

```python
df9.Coordinates[0]
```

```python
df9["Latitude"] = df9["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df9["Longitude"] = df9["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df9
```
